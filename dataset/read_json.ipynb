```python
import pandas as pd
import json

csv_path = "json_dataset_with_changes.csv"
df = pd.read_csv(csv_path)

def count_changes(changes_json):
    try:
        changes_list = json.loads(changes_json)
        return len(changes_list)
    except json.JSONDecodeError:
        return 0

df['num_changes'] = df['changes'].apply(count_changes)

print("Number of changes per entry:")
print(df['num_changes'])

print("\nStatistics:")
print(f"Maximum changes: {df['num_changes'].max()}")
print(f"Minimum changes: {df['num_changes'].min()}")
print(f"Average changes: {df['num_changes'].mean():.2f}")
print(f"Standard deviation of changes: {df['num_changes'].std():.2f}")
```
